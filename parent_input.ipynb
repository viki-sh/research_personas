```python
from dotenv import load_dotenv
import os
from openai import OpenAI
import json

import os
import json
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def generate_child_chat_batch(descriptions, output_file="child_responses.json"):
    system_prompt = (
        "You are an AI that helps generate a sample interaction between a child and a chatbot.\n\n"
        "Given a text description of a child (provided by a parent), your task is to:\n"
        "1. Create a realistic question the child might ask an AI assistant.\n"
        "2. Generate a response the AI assistant would give — appropriate for the child's age, personality, and sensitivities.\n"
        "3. Output the result as a JSON object with the following keys:\n\n"
        "{\n"
        '  "parent_description": "string",\n'
        '  "child_query": "string",\n'
        '  "chatbot_response": "string"\n'
        "}\n\n"
    )

    all_results = []

    for description in descriptions:
        user_prompt = (
            f'Parent input: "{description}"\n\n'
            "Expected Output:"
        )

        response = client.chat.completions.create(
            model="gpt-4.1-nano",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt}
            ],
            temperature=1.0,
            max_tokens=512,
            top_p=1
        )

        try:
            output_json = json.loads(response.choices[0].message.content.strip())
            all_results.append(output_json)
        except Exception as e:
            print(f"Error parsing response for description: {description}\n{e}")
            continue

    with open(output_file, "w") as f:
        json.dump(all_results, f, indent=2)

    return all_results





```
